```python
df_results['solved_boolean'] = df_results.solving_time < 3600
df_results['started_boolean'] = df_results.solving_time != 0
```

```python
select_datasets = ['balance-scale', 'breast-cancer', 'car-evaluation', 'kr-vs-kp', 'monk1', 'soybean-small']
locator = [dataset in select_datasets for dataset in df_results.dataset]
df_results_subset = df_results.loc[locator,:]
```

```python
df_results_subset.pivot_table(values = ['started_boolean', 'solved'], index = ['dataset', 'depth'], columns = 'approach', aggfunc=['sum'])
df_to_latex = df_results_subset.pivot_table(values = ['solved'], index = ['dataset', 'depth'], columns = 'approach', aggfunc=['sum'])
print(df_to_latex.to_latex())
```

```python
pivot_solving_time = df_results_subset.pivot_table(values = ['solving_time'], index = ['dataset', 'depth'], columns = 'approach', aggfunc=['sum'])
```

```python
total_running_time = pivot_solving_time.sum().sum()
total_running_time/(60*60*24)
#Almost 80 days
```
